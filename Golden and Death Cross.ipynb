```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
df = pd.read_excel("/Users/jorgesaldivar/Desktop/BBVA.xlsx")
```

```python
#Creamos las columnas con las medias moviles a 30 y 60 dias
df['M_30']=df['Close'].rolling(window=30).mean()
df['M_60']=df['Close'].rolling(window=60).mean()

#Calculo Puntos de cruce
for i in range(1,1259):
    if df.loc[i-1,'M_60']>df.loc[i-1,'M_30'] and df.loc[i,'M_30']>df.loc[i,'M_60'] :
        df.loc[i,'Cross']= 'Golden'
    elif df.loc[i-1,'M_60']<df.loc[i-1,'M_30'] and df.loc[i,'M_30']<df.loc[i,'M_60'] :
        df.loc[i,'Cross']= 'Dead'
    else :
        df.loc[i,'Cross']= 'NaN'
#Graficamos    
fig, axes = plt.subplots(nrows=2, ncols=1,figsize=(20,10))

df['Close'].plot(ax=axes[0]); axes[0].set_title('Cierre')
df[['M_30','M_60']].plot(ax=axes[1]); axes[1].set_title('Cross Strategy')
```
