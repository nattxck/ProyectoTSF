```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime, date, time, timedelta
import calendar
```

```python
#col_types ={"Idex":int}
#usar read_excel o read_csv
df = pd.read_excel(r"Users/jorgesaldivar/Desktop/BBVA.xlsx")
```

```python
df['V'] = df['Close']
df['Vx'] = df['Close'].rolling(window=10).mean()
df['M'] = (df['V']-df['Vx'])*100

window = 10
df['Promedio de Cierre'] = df['V'].rolling(window).mean()


#Create an "empty" column as placeholder for our /position signals
df['Sugerencia'] = None


for i in range(0,1259):
    if df.loc[i,'M']< -10:
        df.loc[i,'Sugerencia']='Comprar'
        
    elif df.loc[i,'M']>10:
        df.loc[i,'Sugerencia']='Vender'
        
    else :
        df.loc[i,'Sugerencia']='No hay Sugerencia'

df
```
